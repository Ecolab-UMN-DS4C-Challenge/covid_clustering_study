# Create and Save Fips Master Dictionary from https://nrcs.usda.gov

```python
import bs4 as bs
import urllib.request

sauce = urllib.request.urlopen("https://www.nrcs.usda.gov/wps/portal/nrcs/detail/national/home/?cid=nrcs143_013697")
soup = bs.BeautifulSoup(sauce, 'lxml')
table = soup.find('table', {'class': 'data'})
df_fips = pd.read_html(str(table), header=0)[0]
```

```python
df_fips.to_csv(f'{fpath}/data/external/dictionary_maps/fips_usda.csv', index=False)
```

# Set Up Dictionaries for Converting States to Abbreviations

# Pull Data From NY Times, Add FIPS, and Compare to the Assigned FIPS value

```python
df_ny_times = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv')
df_ny_times['UID'] = df_ny_times['county'] + ', ' + df_ny_times['state'].map(states_to_abbr)
df_ny_times
```

```python
df_fips_ny_times = df_ny_times[['fips', 'county', 'state']].dropna().drop_duplicates()
df_fips_ny_times['state'] = df_fips_ny_times['state'].map(states_to_abbr)
df_fips_ny_times['fips'] = df_fips_ny_times['fips'].astype(int)
df_fips_ny_times.columns = ['FIPS', 'Name', 'State']
```

```python
df_fips_master = pd.concat([df_fips, df_fips_ny_times]).drop_duplicates()
df_fips_master.to_csv('../data/external/dictionary_maps/fips_master.csv', index=False)
```
